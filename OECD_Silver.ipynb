```python
# 失業率CSVを読み込む
df_unemployment = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load("Files/Bronze/unemployment.csv")

display(df_unemployment)
```

```python
# GDP成長率CSVを読み込む
df_gdp = spark.read.format("csv")\
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load("Files/Bronze/gdp.csv")

display(df_gdp)
```

```python
# 失業率：必要な列だけ選択してSilverに保存
df_unemployment_silver = df_unemployment.select(
    "REF_AREA",
    "Reference area",
    "TIME_PERIOD",
    "OBS_VALUE",
    "OBS_STATUS"
)

display(df_unemployment_silver)
```

```python
# GDP成長率：必要な列だけ選択
df_gdp_silver = df_gdp.select(
    "REF_AREA",
    "Reference area",
    "TIME_PERIOD",
    "OBS_VALUE",
    "OBS_STATUS"
)

display(df_gdp_silver)
```

```python
# 失業率をSilverテーブルとして保存
df_unemployment_silver.write.format("delta")\
    .mode("overwrite")\
    .option("overwriteSchema", "true")\
    .saveAsTable("silver_unemployment")

# GDP成長率をSilverテーブルとして保存
df_gdp_silver.write.format("delta")\
    .mode("overwrite")\
    .option("overwriteSchema", "true")\
    .saveAsTable("silver_gdp")

print("Silver tables saved!")
```

```python
from pyspark.sql.functions import col

# 失業率：列名を全部変更してSilverテーブルとして保存
df_unemployment_silver = df_unemployment.select(
    col("REF_AREA").alias("CountryCode"),
    col("Reference area").alias("CountryName"),
    col("TIME_PERIOD").alias("Year"),
    col("OBS_VALUE").alias("UnemploymentRate"),
    col("OBS_STATUS").alias("DataStatus")
)

df_unemployment_silver.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("silver_unemployment")

# GDP成長率：列名を全部変更してSilverテーブルとして保存
df_gdp_silver = df_gdp.select(
    col("REF_AREA").alias("CountryCode"),
    col("Reference area").alias("CountryName"),
    col("TIME_PERIOD").alias("Year"),
    col("OBS_VALUE").alias("GDPGrowthRate"),
    col("OBS_STATUS").alias("DataStatus")
)

df_gdp_silver.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("silver_gdp")

print("Silver tables saved!")
```

```python
# dimCountry：国ディメンションを作成
df_dim_country = df_unemployment_silver.select(
    "CountryCode",
    "CountryName"
).distinct()

df_dim_country.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("gold_dimCountry")

display(df_dim_country)
```

```python
# dimYear：年ディメンションを作成
df_dim_year = df_unemployment_silver.select(
    "Year"
).distinct().orderBy("Year")

df_dim_year.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("gold_dimYear")

display(df_dim_year)
```

```python
from pyspark.sql.functions import col

# factEconomic：失業率とGDP成長率を結合
df_fact = df_unemployment_silver.alias("u").join(
    df_gdp_silver.alias("g"),
    (col("u.CountryCode") == col("g.CountryCode")) & 
    (col("u.Year") == col("g.Year")),
    "left"  # クロアチアはGDPデータがないのでleft join
).select(
    col("u.CountryCode"),
    col("u.Year"),
    col("u.UnemploymentRate"),
    col("g.GDPGrowthRate"),
    col("u.DataStatus").alias("UnemploymentStatus"),
    col("g.DataStatus").alias("GDPStatus")
)

df_fact.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("gold_factEconomic")

display(df_fact)
```

```python
%%sql
SELECT 
    f.CountryCode,
    d.CountryName,
    f.Year,
    f.UnemploymentRate,
    f.GDPGrowthRate
FROM gold_factEconomic f
JOIN gold_dimCountry d ON f.CountryCode = d.CountryCode
ORDER BY f.CountryCode, f.Year
```
